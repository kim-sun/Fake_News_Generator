```python
import sys
import pickle
import requests
from datetime import datetime
from bs4 import BeautifulSoup
```

```python
def get_date(news_block_node):
    date_string = news_block_node.find(class_="news_date").string.split('|')[0][2:-1] # you can pass class_ a string, a regular expression, a function, or True
    return(datetime.strptime(date_string, '%Y.%m.%d').strftime('%Y-%m-%d')) # 從2019.03.03開始抓 把2019.03.03轉字串2019-03-03
    # ex:發布2019.03.03 | 00:19 
def get_title(news_block_node):
    return news_block_node.find(class_="news_title").a.string

def get_link(news_block_node):
    return news_block_node.find(class_="news_title").a.get('href')

def get_content(link):
    r = requests.get(link)
    r.encoding = "UTF-8"
    soup = BeautifulSoup(r.text, 'html.parser')
    article_node = soup.find(itemprop='articleBody') #找到這個element
    article = article_node.get_text() # 存純文字的部分
    return article.replace("\n", "")
```

```python
def get_news_info(each_news):
    date  = get_date(each_news)
    title = get_title(each_news)
    link  = get_link(each_news)
    content = get_content(link)
    
    #print(date, title, link)
    
    info = {'date' : date,
            'title': title,
            'link' : link,
            'content': content,
            'tags': 'society'}
    
    return(info)

def get_page_news(page_url): #抓單一頁新聞
    r = requests.get(page_url)
    r.encoding = "UTF-8"

    soup = BeautifulSoup(r.text, 'html.parser')
    news_blocks = soup.find_all(class_="news-list-item clearfix")
    #print(soup)
    #print(news_blocks)
    news = []
    for each_news in news_blocks:
        try:
            news_info = get_news_info(each_news)
        except:
            pass

        news.append(news_info)
    return(news)


def get_new_talk_news(from_page=1, end_page=300, url="https://newtalk.tw/news/subcategory/14/社會"):# 抓很多頁
    print("page_number from {} to {}".format(from_page, end_page -1))
    data = []
    for page_number in range(from_page, end_page):
        print("page_number: {}".format(page_number))
        data = data + get_page_news( url+str(page_number) )# https://newtalk.tw/news/subcategory/2/政治/1 ，此為第一頁
    
    print('done')
    return(data)
```

```python
data = get_new_talk_news(from_page=1, end_page=500)
```
